```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau

# Load and preprocess the CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values to be between 0 and 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# CIFAR-10 class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Data augmentation for training
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)
datagen.fit(X_train)

# Define the CNN model architecture
def create_cnn_model():
    model = Sequential([
        # First Convolutional Block
        Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
        BatchNormalization(),
        Conv2D(32, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Second Convolutional Block
        Conv2D(64, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        Conv2D(64, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Third Convolutional Block
        Conv2D(128, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        Conv2D(128, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Fully Connected Layers
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(10, activation='softmax')
    ])

    return model

# Create and compile the model
model = create_cnn_model()
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Callbacks
checkpoint = ModelCheckpoint(
    'cifar10_best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

# Training parameters
epochs = 25
batch_size = 64

# Train the model
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=batch_size),
    epochs=epochs,
    validation_data=(X_test, y_test),
    steps_per_epoch=len(X_train) // batch_size,
    callbacks=[checkpoint, reduce_lr],
    verbose=1
)

# Load the best model
model.load_weights('cifar10_best_model.h5')

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Accuracy: {test_acc:.4f}")

# Plot training history
plt.figure(figsize=(12, 5))

# Plot training & validation accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot training & validation loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Function to display sample predictions
def display_predictions(model, X, y, n=10):
    # Get random samples
    indices = np.random.choice(range(len(X)), n, replace=False)
    images = X[indices]
    true_labels = np.argmax(y[indices], axis=1)

    # Make predictions
    predictions = model.predict(images)
    predicted_labels = np.argmax(predictions, axis=1)

    # Plot images with predictions
    plt.figure(figsize=(15, 3 * (n // 5 + 1)))
    for i in range(n):
        plt.subplot(n // 5 + 1, 5, i + 1)
        plt.imshow(images[i])

        color = 'green' if predicted_labels[i] == true_labels[i] else 'red'
        title = f"True: {class_names[true_labels[i]]}\nPred: {class_names[predicted_labels[i]]}"

        plt.title(title, color=color)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Show sample predictions
display_predictions(model, X_test, y_test, n=10)

# Class-wise accuracy analysis
def analyze_class_accuracy(model, X, y):
    # Make predictions
    predictions = model.predict(X)
    predicted_labels = np.argmax(predictions, axis=1)
    true_labels = np.argmax(y, axis=1)

    # Calculate accuracy for each class
    class_accuracy = {}
    for i in range(10):
        # Find indices of samples with this true class
        indices = np.where(true_labels == i)[0]
        # Calculate accuracy for this class
        class_acc = np.mean(predicted_labels[indices] == true_labels[indices])
        class_accuracy[class_names[i]] = class_acc

    # Plot class-wise accuracy
    plt.figure(figsize=(12, 6))
    plt.bar(class_accuracy.keys(), class_accuracy.values())
    plt.title('Class-wise Accuracy')
    plt.xlabel('Class')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1.0)
    plt.xticks(rotation=45)
    for i, (name, acc) in enumerate(class_accuracy.items()):
        plt.text(i, acc + 0.01, f"{acc:.3f}", ha='center')
    plt.tight_layout()
    plt.show()

    return class_accuracy

# Analyze class-wise accuracy
class_accuracy = analyze_class_accuracy(model, X_test, y_test)
```
